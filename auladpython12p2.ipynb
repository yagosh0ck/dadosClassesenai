```python
from faker import Faker
from datetime import date
import pandas as pd
```

```python


nomes=[]
dta_nasc=[]

for i in range(20):
  dt = Faker('pt-BR').date_between(start_date='-50y', end_date='-30y')
  dta_nasc.append(dt)


nome=Faker('pt-BR').name()
nomes.append(nome)

print(nomes)
print(dta_nasc)
```

```python
fk={"Nome":[],"Dta":[],"Cidade":[],"Estado":[],"email":[]}
for i in range(20):
  estado=Faker('pt-BR').state()
  nome=Faker('pt-BR').name()
  dt = Faker('pt-BR').date_between(start_date='-50y', end_date='-30y')
  city=Faker('pt-BR').city()
  email=Faker('pt-BR').email()

  fk['email'].append(email)
  fk['Estado'].append(estado)
  fk['Cidade'].append(city)
  fk['Nome'].append(nome)
  fk['Dta'].append(dt) # Append the date to the 'Dta' list

tabela=pd.DataFrame(fk)
display(tabela)
```
