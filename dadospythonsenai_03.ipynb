```python
import pandas as pd
from google.colab import files
from IPython.display import display


colunas=[
    "Nome",
    "Idade",
    "Preferencia(Doce/Salgado)",
    "Cor Favorita",
    "Horas de Sono",
    "Genero Musical"]


tb=pd.DataFrame(columns=colunas)

display(tb)

arquivo_excel="dados_modelo.xlsx"

tb.to_excel(arquivo_excel,index=False,engine="openpyxl")

files.download(arquivo_excel)
print("baixe o arquivo e preenche corretamente com os dados coletados")
```

```python
subir=files.upload()
nome_arquivo=list(subir.keys())[0]
print(f'arquivo{nome_arquivo}')
```

```python
import matplotlib.pyplot as plt
```

```python
dados=pd.read_excel(nome_arquivo,engine="openpyxl")
display(dados)
```

```python
print("cabeçalho")
display(dados.head())
```

```python
print(f'media de idade:{dados["Idade"].mean():.2f} ano')

print(f'media de horas de sono:{dados["Horas de Sono"].mean():.2f} Horas')

print("\n pref. alimentar")
print(dados["Preferencia(Doce/Salgado)"].value_counts())

dados['Genero Musical'].value_counts().plot(kind= 'bar' ,title="Musical")
plt.xlabel("Genero")
plt.ylabel("Numero de aluno")
plt.grid(True)
plt.show()

```

```python
dados['Cor Favorita'].value_counts().plot(kind='bar',title="Cor Favorita")
plt.xlabel("Cor")
plt.ylabel("Quantidade")
plt.grid(True)
plt.show()
```

```python
dados_removidos = dados.drop_duplicates().reset_index(drop=True)
display(dados_removidos)
```
