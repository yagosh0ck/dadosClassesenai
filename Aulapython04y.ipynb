```python
import pandas as pd
from IPython.display import display
```

```python
dados=[

    {"Nome": "Ana", "Idade": 28, "Destinos": "São Paulo","Voo":"KSL00122"},
    {"Nome": "Joao", "Idade": 32, "Destinos": "Rio de Janeiro","Voo":"GNL45226"},
    {"Nome": "Maria", "Idade": None, "Destinos": "Belo Horizonte","Voo":None},
    {"Nome": "Pedro", "Idade": 40, "Destinos": "Salvador", "Voo":"DGY45632"},
    {"Nome": "Julia", "Idade": 22, "Destinos": "Piauí", "Voo":"JLI77777"},
     {"Nome": "Julia", "Idade": 22, "Destinos": "Piauí", "Voo":"JLI77777"},
      {"Nome": "Julia", "Idade": 22, "Destinos": "Piauí", "Voo":"JLI77777"}
]
tb=pd.DataFrame(dados)
display(tb)
```

```python
tb_filtrada=tb.drop_duplicates()
display(tb_filtrada)
```

```python
tb['Idade']=tb['Idade'].fillna(tb['Idade'].median())
tb['Voo']=tb['Voo'].fillna('não informado')
display(tb)
```

```python
import matplotlib.pyplot as plt
```

```python
print (f'media: {tb ["Idade"].mean():.2f}')
print (f'Destinos:')
print (tb["Destinos"].value_counts().plot(kind="bar", title="destinos", color='black'))
plt.ylabel("Quantidade")
plt.xlabel("Destinos")
plt.grid(True)
plt.show()
```
